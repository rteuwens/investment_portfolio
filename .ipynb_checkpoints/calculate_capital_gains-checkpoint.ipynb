```python
import pandas as pd
import numpy as np
from forex_python.converter import CurrencyRates

c = CurrencyRates()
```

```python
# reading our transaction history
df = pd.read_csv('transaction_history.csv')
df.index += 1

# forex_python's CurrencyRates.get_rate requires a datetime object
df['Date'] = pd.to_datetime(df['Date'])

# fetch the fx rates
df['FX to EUR'] = df.apply(lambda row: c.get_rate(row['Currency'], 'EUR', row['Date']), axis=1)

# the metric that matters
df['Cash Movement in EUR'] = df['Quantity'] * df['Price'] * df['FX to EUR'] * -1.00

# printing our dataframe
df
```
